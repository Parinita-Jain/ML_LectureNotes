```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("data2.csv")
df
```

Pand and sklearn ,they both see NaN as missing value.So, we will have to replace all the values with Nan. 

```python
# replace missing value with NaN

df["Python"].replace("missing",np.nan,inplace=True)
df["Machine Learning"].replace("?",np.nan,inplace=True)
df["Age"].replace("missing",np.nan,inplace=True)
df["Age"].replace("?",np.nan,inplace=True)
df["AI"].replace("missing",np.nan,inplace=True)
df["NLP"].replace("?",np.nan,inplace=True)
df["Cv"].replace("?",np.nan,inplace=True)
df["Cv"].replace("missing",np.nan,inplace=True)
```

```python
df.head()
```

We want all the rows, and except name col, Iwant all the col and we need to change the datatype. 

```python
df.iloc[:,1:]=df.iloc[:,1:].astype("float64")
```

```python
from sklearn.impute import SimpleImputer ,KNNImputer # these are the 2 different types of imputer.
```

```python
si=SimpleImputer(missing_values=np.nan,strategy="mean") # si is the object of class SimpleImputer,by default strategy
# is mean
# we r simply telling,wherever u see NaN, fill it with mean.
```

```python
df.iloc[:,1:]
```

```python
df.iloc[:,1:]=si.fit_transform(df.iloc[:,1:]) # fit does calc and get mean value and transform, the data.
# no inplace parameter is there.
```

U can use KNNImputer too, insted of SimpleImputer. When it comes to missing values,we can fill with mean and median.If the col is skewed any right or left,and numerical data in it,then go with median value. In case of no skewness i.e. normal distribution, go with mean.
If the categorical data is missing,go with mode
