```python
import pandas as pd
import numpy as np
import networkx as nx

import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("routes.csv")
df.head()
```

```python
g=nx.Graph() # fromthe nx package creating an object of graph

# now we just need source airport and destination airport,using which we will create graph
g=nx.from_pandas_edgelist(df,source='source airport',target='destination airport')
```

```python
# now this graph has so many nodes, so many edges,so its difficult to visualize
# but we can still look into it

print(g.nodes)
print(g.edges)

```

```python
dc=nx.degree_centrality(g)  # getting degree centrality

print("Degree Cetrality : ",dc) # no.ofdirect connections

# so,this gives ans towhich airport has direct connections with other airport

#but we donot want to see degree centrality of each node,buttop few with highest degree centrality..then-- 

# So,we need to sort this degreeCentrality which is in dictionary format

sorted_dc={k: v for k,v in sorted(dc.items(),key=lambda item:item[1],reverse=True) } # sort the dict descending 


print("Sorted DC : ",sorted_dc)

# airport AMS is having highest_degee_centrality i.e.if by any chance u goto AMS airport,your flight will
# be a direct flight
```

```python
# then getting closeness centrality

closeness=nx.closeness_centrality(g)

# now checking closenes centrality..i.e. which airport is very close to  other airports
print("\nClosenessCentrality : ",closeness)

sorted_c={k: v for k,v in sorted(closeness.items(),key=lambda item:item[1],reverse=True) }
print("\nSortedCC",sorted_c)

# airport FRA is close to most of the airports. 
```

```python
# then getting betweenness centrality

betweenness=nx.betweenness_centrality(g)

# now checking betweenness centrality..
print("\nClosenessCentrality : ",betweenness)

sorted_c={k: v for k,v in sorted(betweenness.items(),key=lambda item:item[1],reverse=True) }
print("\nSortedCC",sorted_c)

# This tells ANC is an international hub ..i.e. most of the flights which are going from source to destination
# has a layover at ANC
```
