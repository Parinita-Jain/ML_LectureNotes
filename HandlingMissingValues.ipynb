```python
df.dropna(how="all") #remove all those rows where all the cols has missing value
# drop only those rows where all the three cols are missing.
# Now,in our dataset,wehave no rowwhich has all the values missing.
```

Now,wehave another parameter inside drop, which is subset.

```python
df.dropna(how="all", subset=["Python" ,"Machine Learning"])# check NaN values in python and ml col.
# the records which have values missing in both the cols,will be dropped.
```

But,I want all my rows. So I can use fillna(), which fills the value with some value.

```python
df.fillna(5)
```

But this is not right.I should not corrupt my data with random data. So I can use values like mean,mode,etc. 

```python
pmean=df["Python"].mean()
```

```python
mmean=df["Python"].mean()
```

```python
df["Python"].fillna(pmean,inplace=True)
df["Machine Learning"].fillna(mmean,inplace=True)
```
