```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules 
```

```python
df=pd.read_csv("GroceryStoreDataSet.csv",names=['products'],sep=',')
df.head()
```

```python
data=list(df['products'].apply(lambda x:x.split(",")))
data
```

```python
# we are changing format of data
#maggi sauce m s c br but ch
#            1 1 0 0   0   0
#maggi chips 1 0 1 0   0   0
# i.e. converting into one hot encoding.
# the algorithm that we will run ,expect input in the numeric format.
# So above we got list of products.Now we will convert products into this format.
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
encoded_data=te.fit(data).transform(data)
encoded_data=encoded_data.astype('int')

df=pd.DataFrame(encoded_data,columns=te.columns_)
df.astype("int")

df.head()
```

```python
df.shape
```

```python
scores=apriori(df,min_support=0.2,use_colnames=True) # min_support is threshold
# if u have support value=0.2,out of 20 trxs what value u will put in numerator toget 0.2 threshold
#4/20=0.2
scores
```

```python
scores_ar=association_rules(scores,metric='confidence',min_threshold=0.6) # u can even go for lift value.or supprt value.
scores_ar
```
