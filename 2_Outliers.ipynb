# Handling missing with SimpleImputer

```python
features["normalized-losses"].replace("?", np.nan, inplace=True)
features["horsepower"].replace("?", np.nan, inplace=True)
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values=np.nan, strategy="mean")
features[["normalized-losses", "horsepower"]] = si.fit_transform(features[["normalized-losses", "horsepower"]])
```

```python
features
```

# Outliers

Drawwhiskers box plot, whichis in sns library,to get outliers.

```python
sns.boxplot(data=features, x=target)
plt.grid()
```

There is no unit over here of price i.e. rupees,pound , etc. Currently ur box plot is treating every value above 30000 as outlier,but do u think all the cars price in the world arein the range 5000-30000. When dealing with real world problem we need to check the outlier in price by considering brand alwasys, as different brands will have different starting and stopping price range. Therefore we need to  look for price w.r.t. brand. Now,make col gives brands. So to get outliers in brand col w.r.t brand Merceds benz and jaguar are above pricerangeof 30000 which were earlier treated as outlier. Therefore, in real world always create whisker boxplot with 2 variables. All the brands have different starting and ending range. The highest is considered as outlier.

```python
sns.boxplot(data=features, x=target, y="make")
plt.grid()
```

Now,if u see dodge ,there are outliers. Because 2nd line ofgrid is 10000, therefore we can have price>10000 ,and wee get the index as 29.So we will drop this 29th index.

```python
features[(features.make=="dodge")&(target>10000)]
```
